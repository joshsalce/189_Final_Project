```python
import pandas as pd
import statistics as st
```

```python
# Read in participants info file from OpenNeuro dataset
articipants = pd.read_csv("participants.tsv", sep="\t")
participants
```

```python
# Calculate average age of all participants
mean_age = round(sum(participants.age.dropna()) / len(participants.age.dropna()), 0)
mean_age
# 3, 9, 19, 21, 23
```

```python
# Calculate median age of all participants

median_age = st.median(participants.age.dropna())
median_age
```

This script allows us to find the parameters by which we will select data from our participants to analyze. Since we are not going to analyze all of our data, we want to select participants who fall close to the mean and median of the age distribution for our participants. We also want to select two participants from each category of 'gender' and 'group.' We aim to select two datasets each from  of our four main participants (male and novice, female and novice, male and expert, female and expert).

Given this criteria, we will select participants 3, 9, 19, 21, 23.
